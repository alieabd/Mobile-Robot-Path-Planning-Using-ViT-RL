```python
def r_t(pose1,pose2):
    
    theta = atan((pose2[0][1]-pose1[0][1])/(pose2[0][0]-pose1[0][0])) - pose1[1][2]
    
    R = ((pose2[0][1]-pose1[0][1])**2 + (pose2[0][0]-pose1[0][0])**2)**0.5
    
    return [R,theta]

r_t_data = []
```

```python
episode_range = [[i for i in range(0,100)],[i for i in range(100,200)],[i for i in range(200,300)],
                    [i for i in range(300,400)],[i for i in range(400,500)],[i for i in range(500,600)],
                                [i for i in range(600,700)],[i for i in range(700,800)]]

for episode in range(8):
    for i in range(99):
        r_t_data.append(r_t(pose[episode_range[episode][i]],pose[episode_range[episode][i+1]]))
```
