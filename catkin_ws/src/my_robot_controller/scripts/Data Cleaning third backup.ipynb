```python
from math import atan
def r_t(pose1,pose2):
    
    theta = atan((pose2[1]-pose1[1])/(pose2[0]-pose1[0])) - pose1[2]
    
    if np.isnan(theta):
        theta = 0
    
    R = ((pose2[1]-pose1[1])**2 + (pose2[0]-pose1[0])**2)**0.5
    
    return [R,theta]
```

```python
Current_pose = np.delete(current_pose,element_to_delete,axis = 0)
```

```python
def episode_lengths(parameters):
    lengths = []
    prev_end = 0
    for end in parameters:
        length = end - prev_end
        lengths.append(length)
        prev_end = end
    return lengths


episode_lengths = episode_lengths(Episode)
r_t_data = []
for last,length in zip(Episode,episode_lengths):
    for i in range(int(length)):
        #print(f'{last - length + i+1} ==> {last} $$ {Current_pose[int(last - length + i+1 - 1)][0]} - {Current_pose[int(last)-1][0]} ==> {r_t(Current_pose[int(last - length + i+1 - 1)][0],Current_pose[int(last)-1][0])}')
        r_t_data.append(r_t(Current_pose[int(last - length + i+1 - 1)][0],Current_pose[int(last)-1][0]))
```
