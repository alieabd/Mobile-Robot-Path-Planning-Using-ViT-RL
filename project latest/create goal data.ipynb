```python
import pandas as pd
file = pd.read_csv("goal.csv",header=None)
from math import atan
```

```python
file
```

```python
def r_t(pose2,pose1 = [[0.0013064784228448718,0.0032157046744297046]]):

    theta = atan((pose2[1]-pose1[0][1])/(pose2[0]-pose1[0][0]))

    R = ((pose2[1]-pose1[0][1])**2 + (pose2[0]-pose1[0][0])**2)**0.5

    return [R,theta]


r_t_data = []
```

```python
for i in range(70):
    pose2 = file.loc[i]
    data = r_t(pose2)
    r_t_data.append(data)
```
